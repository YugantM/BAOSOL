```python
import json
import pandas as pd
```

```python
def flatten_json(filepath):
    # opening file and loading JSON
    file = open(filepath,)  
    data = json.load(file)

    # creating a dataframe to store the result
    columns = "item_name, rel_answer, rel_objection, text, count, rel_count, type"
    result_frame = pd.DataFrame(columns= columns.split(","))

    # row counter
    c = 0


    for each_item in data['items']:

        # initial three items for each row are obtained directly from the repective key
        fixed = [each_item['name'],each_item['rel_answer'],each_item['rel_objection']]


        for x in each_item['answers']:

            # obtaining data from each item in answers
            result_frame.loc[c] = fixed+ [x['text'],x['count'],x['rel_count'],'Answer']
            c+=1

        for x in each_item['objections']:

            # obtaining data from each item in objections
            result_frame.loc[c] = fixed+ [x['text'],x['count'],x['rel_count'],'Objection']
            c+=1


    return result_frame
```

```python
flatten_json('./input.json')
```
